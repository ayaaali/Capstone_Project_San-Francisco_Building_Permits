## Filling The Missing Values

```python
def fill_missing_values_all_columns(df, group_columns):
    df_copy = df.copy()

    for col in df_copy.columns.difference(group_columns):
        # Fill NaNs with mode within each group
        df_copy[col] = df_copy.groupby(group_columns)[col].transform(
            lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else x)
        )
    return df_copy


# Example usage
group_columns = ['Street Name', 'Block']
df_filled = fill_missing_values_all_columns(df, group_columns)

# Check missing values sorted by count
missing_values = df_filled.isnull().sum().sort_values(ascending=False)
print(missing_values)
```

```python
import matplotlib.ticker as mticker

# Ensure date column is datetime
df['Permit Creation Date'] = pd.to_datetime(df['Permit Creation Date'], errors='coerce')
df['Year'] = df['Permit Creation Date'].dt.year

# Calculate average days to issue for each permit type
mean_days_permit_type = df.groupby('Permit Type Definition')['Days_to_Issue'].mean().reset_index()

# Get the top 3 slowest permit types
top3_permit_types = mean_days_permit_type.sort_values(by='Days_to_Issue', ascending=False).head(3)['Permit Type Definition']

# Filter only those permit types
df_top3 = df[df['Permit Type Definition'].isin(top3_permit_types)]

# Group by Year and Permit Type for average days
avg_days_yearly = df_top3.groupby(['Year', 'Permit Type Definition'])['Days_to_Issue'].mean().reset_index()

# Plot each permit type separately
plt.figure(figsize=(15, 12))

for i, permit in enumerate(top3_permit_types, 1):
    plt.subplot(3, 1, i)  # 3 rows, 1 column, index i
    temp = avg_days_yearly[avg_days_yearly['Permit Type Definition'] == permit]
    sns.lineplot(x='Year', y='Days_to_Issue', data=temp, marker='o', color='skyblue')
    plt.title(f"Average Approval Days Over Time for {permit}")
    plt.xlabel("Year")
    plt.ylabel("Average Days to Issue")
    plt.grid(True)
    
    # Force x-axis to show only integer years
    plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```

# Task 9: How do costs differ across types of permits?

```python
import matplotlib.ticker as mtick

# Create dataframe for plotting
df_cost = df[['Permit Type Definition', 'Estimated Cost']].dropna()

# Calculate median cost per permit type
median_cost = df_cost.groupby('Permit Type Definition')['Estimated Cost'].median().sort_values(ascending=False).reset_index()

# Plot horizontal bar chart
plt.figure(figsize=(12,6))
ax = sns.barplot(data=median_cost, y="Permit Type Definition", x="Estimated Cost", color="skyblue")
plt.title("Median Estimated Cost by Permit Type")
plt.xlabel("")
plt.ylabel("Permit Type")

# Format x-axis with K/M (for annotation purposes)
def format_currency(x, pos):
    if x >= 1e6:
        return f'${x/1e6:.1f}M'
    elif x >= 1e3:
        return f'${x/1e3:.0f}K'
    else:
        return f'${int(x)}'

ax.xaxis.set_major_formatter(mtick.FuncFormatter(format_currency))

# Show values at the end of each bar
for p in ax.patches:
    value = p.get_width()
    if value >= 1e6:
        label = f'${value/1e6:.1f}M'
    elif value >= 1e3:
        label = f'${value/1e3:.0f}K'
    else:
        label = f'${int(value)}'
    ax.annotate(label,
                (value, p.get_y() + p.get_height() / 2.),
                ha='left', va='center', fontsize=10, color='black')

# Remove x-axis
ax.xaxis.set_visible(False)

plt.show()
```
