```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Building_Permits_1.csv",low_memory= False) # Load CSV
```

```python
df["Permit Number"] = df["Permit Number"].astype(str)
df
```

```python
date_columns = ['Permit Creation Date', 'Current Status Date', 'Filed Date', 'Issued Date',
                 'Completed Date', 'First Construction Document Date', 'Permit Expiration Date']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce') # Converting to datetime format
```

```python
# Create a new column for Challenge 1

df['Days_to_Issue'] = (df['Issued Date'] - df['Filed Date']).dt.days
```

```python
fire_only_permit_counts = df['Fire Only Permit'].value_counts()
structural_notification_counts = df['Structural Notification'].value_counts()
site_permit_counts = df['Site Permit'].value_counts()

print(fire_only_permit_counts)
print(structural_notification_counts)
print(site_permit_counts)
```

```python
# Columns have only 'Y' = 'Yes' option. Let's transform it to boolean format and get rid of the nulls

df['Fire Only Permit'] = df['Fire Only Permit'].map({'Y': True, np.nan: False})
df['Structural Notification'] = df['Structural Notification'].map({'Y': True, np.nan: False})
df['Site Permit'] = df['Site Permit'].map({'Y': True, np.nan: False})

print(df['Fire Only Permit'].value_counts())
print(df['Structural Notification'].value_counts())
print(df['Site Permit'].value_counts())
```

```python
missing_values = df.isnull().sum()
missing_values = missing_values.sort_values(ascending=False)
missing_values
```

## Filling The Missing Values

```python
def fill_missing_values_all_columns(df, group_columns):
    df_copy = df.copy()

    for col in df_copy.columns.difference(group_columns):
        # Fill NaNs with mode within each group
        df_copy[col] = df_copy.groupby(group_columns)[col].transform(
            lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else x)
        )
    return df_copy


# Example usage
group_columns = ['Street Name', 'Block']
df_filled = fill_missing_values_all_columns(df, group_columns)

# Check missing values sorted by count
missing_values = df_filled.isnull().sum().sort_values(ascending=False)
print(missing_values)
```

# Task 1: Permits Analysis

```python
permit_type_def_counts = df['Permit Type Definition'].value_counts()
print(permit_type_def_counts)
```

```python
# What types of permits are most and least common in San Francisco?
plt.figure(figsize=(12, 6))
count_plot = sns.countplot(
    x='Permit Type Definition',
    data=df,
    order=df['Permit Type Definition'].value_counts().index,
    color='skyblue'  # light blue color
)
plt.title('Permit Type Definition Counts')
plt.xlabel('Permit Type Definition')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

# Annotate bars with counts
for p in count_plot.patches:
    count_plot.annotate(f'{p.get_height()}', 
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
summary = df.groupby("Permit Type Definition")["Days_to_Issue"].agg(["mean","median","count"]).reset_index()
print(summary)
```

```python
# Average Days to Issue by Permit Type (horizontal)
mean_days_permit_type = df.groupby('Permit Type Definition')['Days_to_Issue'].mean().reset_index()
mean_days_permit_type = mean_days_permit_type.sort_values(by='Days_to_Issue')

plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(
    y='Permit Type Definition',  # flip axes
    x='Days_to_Issue',
    data=mean_days_permit_type,
    color='skyblue'
)

plt.title('Average Days to Issue for Each Permit Type')
plt.xlabel('Average Days to Issue')
plt.ylabel('Permit Type Definition')

# Annotate bars with values
for p in bar_plot.patches:
    width = p.get_width()  # horizontal bar, use width
    bar_plot.annotate(f'{width:.0f}',  # no decimals
                      (width, p.get_y() + p.get_height() / 2.),
                      ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.show()
```

```python
df.dropna(subset=['Days_to_Issue'], inplace=True) 

permit_type_counts = df['Permit Type Definition'].value_counts().reset_index()
permit_type_counts.columns = ['Permit Type Definition', 'Count']

mean_days_permit_type = df.groupby('Permit Type Definition')['Days_to_Issue'].mean().reset_index()
mean_days_permit_type = mean_days_permit_type.sort_values(by='Days_to_Issue')
mean_days_permit_type.columns = ['Permit Type Definition', 'Avg Days to Issue']

result_df = pd.merge(permit_type_counts, mean_days_permit_type, on='Permit Type Definition')
print(result_df)
```

# Task 2: Best day for DBI Visit

```python
# whether the day of the week when a permit is filed has any effect on how quickly it gets approved (issued)
df['Weekday'] = df['Filed Date'].dt.day_name()

average_days_per_weekday = df.groupby('Weekday')['Days_to_Issue'].mean().reset_index()
best_day = average_days_per_weekday.loc[average_days_per_weekday['Days_to_Issue'].idxmin()]

plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Weekday', y='Days_to_Issue', data=average_days_per_weekday, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'], color='skyblue')

for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Average Days to Issue for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Days to Issue');
```

# Task 3: Number of Floors

```python
# Most common number of stories in buildings applying for permits
plt.figure(figsize=(10, 6))
hist_plot = sns.histplot(df['Number of Existing Stories'], bins=20, color='skyblue', kde=False)  # bars only

plt.title('Histogram for Number of Existing Stories')
plt.xlabel('Number of Existing Stories')
plt.ylabel('Frequency')

# Get all bar heights
heights = [p.get_height() for p in hist_plot.patches]

# Find indices of top 3 tallest bars
top3_indices = sorted(range(len(heights)), key=lambda i: heights[i], reverse=True)[:3]

# Annotate only top 3 bars
for i in top3_indices:
    p = hist_plot.patches[i]
    height = p.get_height()
    hist_plot.annotate(f'{int(height)}',
                       (p.get_x() + p.get_width() / 2., height),
                       ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
df['Permit Creation Date'] = pd.to_datetime(df['Permit Creation Date'], errors='coerce')
df['Year'] = df['Permit Creation Date'].dt.year

# Group by year
permits_per_year = df.groupby('Year')['Permit Number'].count()

# Plot
plt.figure(figsize=(10,6))
permits_per_year.plot(kind='line', marker='o', color='skyblue')
plt.title("Permit Applications Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Permits")
plt.grid(True);
```
